```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_excel("ABC Company.xlsx")
df.head()

# Replace Height
df['Height'] = np.random.randint(150, 181, len(df))

# Count employees in each team
team_counts = df["Team"].value_counts()

# Total employees
total_employees = len(df)

# Calculate percentage
team_percentages = (team_counts / total_employees) * 100

# Combine count and percentage in one table
team_distribution = pd.DataFrame({
    "Employee Count": team_counts,
    "Percentage (%)": team_percentages.round(2)
})

print(team_distribution)

plt.figure(figsize=(12,6))

team_counts.plot(kind="bar")

plt.title("Employee Distribution by Team")
plt.xlabel("Team")
plt.ylabel("Number of Employees")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()


# Count employees in each position
position_counts = df["Position"].value_counts()

print(position_counts)

plt.figure(figsize=(12,6))
position_counts.plot(kind="bar", color="gold")

plt.title("Employee Distribution by Position")
plt.xlabel("Position")
plt.ylabel("Number of Employees")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Create age bins
bins = [18,25,30,35,40,45,50]
labels = ["18-25","26-30","31-35","36-40","41-45","46-50"]

df["Age Group"] = pd.cut(df["Age"], bins=bins, labels=labels, include_lowest=True)

age_counts = df["Age Group"].value_counts().sort_index()

print(age_counts)
print("Predominant Age Group:", age_counts.idxmax())

plt.figure(figsize=(6,6))

age_counts.plot(kind="pie", autopct="%1.1f%%")

plt.title("Age Group Distribution")
plt.ylabel("")
plt.show()

# Salary Expenditure Analysis

# ----- Team Salary -----
team_salary = df.groupby("Team")["Salary"].sum().sort_values(ascending=False)
print("Total Salary by Team:")
print(team_salary)
print("Team with Highest Salary:", team_salary.idxmax())

plt.figure(figsize=(12,6))

team_salary.sort_values().plot(kind="barh", color="teal")

plt.title("Total Salary by Team")
plt.xlabel("Total Salary")
plt.ylabel("Team")
plt.xticks(rotation=90)

plt.tight_layout()
plt.show()

# ----- Position Salary -----
position_salary = df.groupby("Position")["Salary"].sum().sort_values(ascending=False)

print("\nTotal Salary by Position:")
print(position_salary)
print("Position with Highest Salary:", position_salary.idxmax())

plt.figure(figsize=(8,5))
position_salary.plot(kind="bar", color="blue")
plt.title("Total Salary by Position")
plt.xlabel("Position")
plt.ylabel("Total Salary")
plt.xticks(rotation=45)
plt.show()


# Correlation between Age and Salary

# Calculate correlation
correlation = df["Age"].corr(df["Salary"])

print("Correlation between Age and Salary:", correlation)

# Scatter Plot (Visual Representation)
plt.figure(figsize=(8,5))
plt.scatter(df["Age"], df["Salary"], alpha=0.6)

plt.title("Age vs Salary")
plt.xlabel("Age")
plt.ylabel("Salary")
plt.grid(True)
plt.show()

```

**Data Story**

The analysis shows that employee distribution is uneven across teams and positions, indicating that certain departments play a more dominant role in the organization. The age group analysis reveals that the majority of employees fall within a specific age range, suggesting a workforce concentrated in a particular career stage.

Salary expenditure analysis highlights that certain teams and positions account for the highest total salary spending, reflecting their importance and responsibility level within the company. The correlation between age and salary suggests whether compensation tends to increase with age and experience.

Overall, the data reveals clear patterns in workforce structure, salary allocation, and demographic trends within the organization.
